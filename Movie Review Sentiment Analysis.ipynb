```python
#Import Libraries
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

#Declare positive, negative and neutral word counters
positiveWordCount = 0
negativeWordCount = 0
neutralWordCount = 0

#Count positive and negative words in the review and calculate positivity and negativity
def control(review):
    review_tokenized = word_tokenize(review)

    positive_counter = 0
    negative_counter = 0

    for word in review_tokenized:
        if word not in ingStopWords:
            for positive_word in pw:
                if(word == positive_word.strip()):
                    positive_counter = positive_counter +1
                    positiveWordCount = positiveWordCount +1
                    break
        else:
            neutralWordCount = neutralWordCount +1


    for word in review_tokenized:
        if word not in ingStopWords:
            for negative_word in nw:
                if(word == negative_word.strip()):
                    negative_counter = negative_counter +1
                    negativeWordCount = negativeWordCount +1
                    break
        else:
            neutralWordCount = neutralWordCount +1            
    
    
    if(positive_counter >= negative_counter):
        return 1
    else:
        return 2

#Import positive words dataset file and save in a variable
f = open("Positive Words.txt",'r')
pw = f.readlines()

#Import negative words dataset file and save in a variable
f = open("Negative Words.txt",'r')
nw = f.readlines()

#Import IMDB movie reviews dataset file and save in a variable
f = open("Test.txt",'r')
reviews = f.readlines()

#Take stop words in english and save in a variable
ingStopWords = set(stopwords.words("english"))

#Declare counter to know which review is processing
count = 0

#Declare positive review and negative review counters
positive_counter = 0
negative_counter = 0

#Process reviews, find which review is negative or positive and total number of positive and negative reviews
for review in reviews:
    count = count + 1
    
    if control(review) == 1:
        positive_counter = positive_counter + 1
        print(str(count) + ". review is positive")
    else:
        negative_counter = negative_counter + 1
        print(str(count) + ". review is negative")

#Display which review is negative or positive and total        
print("\nNumber of negative reviews: " + str(negative_counter))
print("Number of positive reviews: " + str(positive_counter))
```

```python
#Plot a pie chart

#Import Libraries
import matplotlib.pyplot as plt

#Data of plot
labels = 'Positive Reviews', 'Negative Reviews'
sizes = [positive_counter, negative_counter]
colors = ['lightcoral', 'lightskyblue']
explode = (0,0)

#Create plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.2f%%', shadow=True, startangle=180)

#Show plot
plt.axis('equal')
plt.show()
```
