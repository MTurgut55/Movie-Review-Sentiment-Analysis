### Description:
##### These days there are a lot of local and foreign movies. Many databases have been developed to have opinion about these movies. Internet Movie Database (Imdb) is most known of these databases. In this database, audiences comment about movies and then they give points to movies a number between one and ten. Mean of all points shows the movie’s point. But many times points disaccord comments of same movies or points aren’t reflect the comments. So one audience has to two act: looking points and reading comments.  In addition while audiences are reading comments, they can read spoiler comments about movie. Aim of this project is making an analysis of comment on Imdb and understanding that how movie evaluated without reading comments. In this project, if positive comment are more than negative comments, audience can say “Movie gets positive comments.” . If there is a reverse situation, audience can say “Movie gets negative comments.”. Owing to this project, without reading comment about movies audiences can understand the movie is good or bad.

```python
#Import Libraries
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

#Declare positive, negative word counters
positiveWordCount = 0
negativeWordCount = 0
neutralWordCount = 0
#Count positive and negative words in the review and calculate positivity and negativity
def control(review):
    review_tokenized = word_tokenize(review)

    positive_counter = 0
    negative_counter = 0
    neutralWordCount = 0

    global positiveWordCount
    global negativeWordCount
    
    
    for word in review_tokenized:
        if word not in ingStopWords:
            for positive_word in pw:
                if(word == positive_word.strip()):
                    positive_counter = positive_counter +1
                    break
            for negative_word in nw:
                if(word == negative_word.strip()):
                    negative_counter = negative_counter +1
                    break   
        else:
            neutralWordCount = neutralWordCount +1
    
    positiveWordCount = positiveWordCount + positive_counter
    negativeWordCount = negativeWordCount + negative_counter
    
    if(positive_counter >= negative_counter):
        return 1
    else:
        return 2

#Import positive words dataset file and save in a variable
f = open("Positive Words.txt",'r')
pw = f.readlines()

#Import negative words dataset file and save in a variable
f = open("Negative Words.txt",'r')
nw = f.readlines()

#Import IMDB movie reviews dataset file and save in a variable
f = open("Test.txt",'r')
reviews = f.readlines()

#Take stop words in english and save in a variable
ingStopWords = set(stopwords.words("english"))

#Declare counter to know which review is processing
count = 0

#Declare positive review and negative review counters
positive_counter = 0
negative_counter = 0

#Process reviews, find which review is negative or positive and total number of positive and negative reviews
for review in reviews:
    count = count + 1
    
    if control(review) == 1:
        positive_counter = positive_counter + 1
        print(str(count) + ". review is positive")
    else:
        negative_counter = negative_counter + 1
        print(str(count) + ". review is negative")

#Display which review is negative or positive and total        
print("\nNumber of negative reviews: " + str(negative_counter))
print("Number of positive reviews: " + str(positive_counter))
```

```python
#Plot a pie chart

#Import Libraries
import matplotlib.pyplot as plt

#Data of plot
labels = 'Positive Reviews', 'Negative Reviews'
sizes = [positive_counter, negative_counter]
colors = ['lightcoral', 'lightskyblue']
explode = (0,0)

#Create plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.2f%%', shadow=False, startangle=225)

#Show plot
plt.axis('equal')
plt.show()
```

```python
#Importing Libraries
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#Data of plot
objects = ('Positive Words', "Negative Words", "Neutral Words")
y_pos = np.arange(len(objects))
values = [positiveWordCount, negativeWordCount, neutralWordCount]

#Create plot
plt.bar(y_pos, values, color = ['blue', 'red', 'gray'], align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number of words')
plt.title('Positive, Negative and Neutral Words Bar Chart')

#Show plot
plt.show()
```
